### Salary parsing

```python
df=df[df['Salary Estimate']!= '-1']
```

```python
df.shape
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary
```

```python
minus_kd=salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python
df['hourly']=df['Salary Estimate'].apply(lambda x: 1 if 'perhour' in x.lower() else 0)
```

```python
df['employer_provided']=df['Salary Estimate'].apply(lambda x: 1 if 'employer provided' in x.lower() else 0)
```

```python
min_hr = minus_kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary',''))
```

```python
min_hr
```

```python
df['min_salary']=min_hr.apply(lambda x: int(x.split('-')[0].replace(':','')))
```

```python
df['min_salary']
```

```python
df['max_salary']=min_hr.apply(lambda x: int(x.split('-')[1].replace(':','')))
```

```python
df['avg_salary']=(df['min_salary']+df['min_salary'])/2
```

### company name text only

```python
df['company_txt']=df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-3],axis=1)
```

```python
df
```

### State Field 

```python
df['job_state']=df['Location'].apply(lambda x: x.split(',')[1])
```

```python
df
```

```python
df.job_state.value_counts()
```

```python
df['same_state']=df.apply(lambda x: 1 if x.Location == x.Headquarters else 0,axis=1)
```

### Age of Company 

```python
df['age'] = df['Founded'].apply(lambda x: x if x<1 else 2023-x)
```
